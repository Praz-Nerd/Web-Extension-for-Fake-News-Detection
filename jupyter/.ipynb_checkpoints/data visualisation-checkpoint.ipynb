```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
news_df = pd.read_csv('../kaggle_dataset/news.csv')
liar_df = pd.read_csv('../liar_dataset/liar.csv')


fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot for Kumar Dataset
kumar_counts = news_df['label'].value_counts()
bars1 = kumar_counts.plot(kind='bar', ax=axes[0], color='skyblue', edgecolor='black')
axes[0].set_title('Fake vs. True Counts: Kumar Dataset')
axes[0].set_xlabel('Label')
axes[0].set_ylabel('Count')
axes[0].bar_label(bars1.containers[0])

# Plot for LIAR Dataset
liar_counts = liar_df['label'].map({0: 'REAL', 1: 'FAKE'}).value_counts()
bars2 = liar_counts.plot(kind='bar', ax=axes[1], color='salmon', edgecolor='black')
axes[1].set_title('Fake vs. True Counts: LIAR Dataset')
axes[1].set_xlabel('Label')
axes[1].set_ylabel('Count')
axes[1].bar_label(bars2.containers[0])

plt.tight_layout()
plt.show()
```
